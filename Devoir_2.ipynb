# Question 1
Soit un stratifié graphite-époxyde $[±60/0]_S$ subissant une rotation θ dans le plan par rapport à
l’axe x.

## a)  Déterminer les propriétés équivalentes $\bar{E}_{x}$, $\bar{E}_{y}$, $\bar{G}_{xy}$, $\bar{\nu}_{xy}$ et $\bar{\nu}_{yx}$ pour $\theta=0$

```python
from composite import CompositeType
from laminate import Laminate
import numpy as np

laminate = Laminate(thetas=[60, -60, 0, 0, -60, 60], h=0.150, composite_type=CompositeType.Graphite_Epoxy)
laminate.display_effective_properties()
```

## b) Calculez et tracez les coefficients de la matrice $A$ ($A_{11}$, $A_{12}$, $A_{22}$, $A_{16}$, $A_{66}$) en fonction de $\theta$ pour -90º < $\theta$ < 90º 

```python
import matplotlib.pyplot as plt

fig_a_11 = plt.figure()
fig_a_12 = plt.figure()
fig_a_22 = plt.figure()
fig_a_16 = plt.figure()
fig_a_66 = plt.figure()
a_11_vals, a_12_vals, a_22_vals, a_16_vals, a_66_vals = [], [], [], [], []
theta_vals = range(-90, 91, 1)
# [60, -60, 0, 0, -60, 60]

for theta in theta_vals:
    thetas = np.array([60, -60, 0, 0, -60, 60]) + theta
    laminate = Laminate(thetas=thetas, h=0.150, composite_type=CompositeType.Graphite_Epoxy)
    a_matrix = np.round(laminate.a_matrix / 1e6, 4)
    
    a_11_vals.append(a_matrix[0, 0])
    a_12_vals.append(a_matrix[0, 1])
    a_22_vals.append(a_matrix[1, 1])
    a_16_vals.append(a_matrix[0, 2])
    a_66_vals.append(a_matrix[2, 2])

fig_a_11.gca().plot(theta_vals, a_11_vals, 'r-')
fig_a_12.gca().plot(theta_vals, a_12_vals, 'r-')
fig_a_22.gca().plot(theta_vals, a_22_vals, 'r-')
fig_a_16.gca().plot(theta_vals, a_16_vals, 'r-')
fig_a_66.gca().plot(theta_vals, a_66_vals, 'r-')

fig_a_11.gca().set_title('A11')
fig_a_11.gca().set_xlabel('Theta (º)')
fig_a_11.gca().set_ylabel('GPa')
fig_a_12.gca().set_title('A12')
fig_a_12.gca().set_xlabel('Theta (º)')
fig_a_12.gca().set_ylabel('GPa')
fig_a_22.gca().set_title('A22')
fig_a_22.gca().set_xlabel('Theta (º)')
fig_a_22.gca().set_ylabel('GPa')
fig_a_16.gca().set_title('A16')
fig_a_16.gca().set_xlabel('Theta (º)')
fig_a_16.gca().set_ylabel('GPa')
fig_a_66.gca().set_title('A66')
fig_a_66.gca().set_xlabel('Theta (º)')
fig_a_66.gca().set_ylabel('GPa')
```

# c) Quelle conclusion pouvez-vous tirer sur le comportement de ce stratifié? 

Il semble que la matrice A est indépendante de $\theta$. À noter également que $A_{11}$ = $A_{22}$,  $A_{16}=A_{26}$ et $A_{66}=(A_{11} - A_{12})/2$, ce qui, après recherche, semble être le comportement d'un stratifié isotrope.
